```python
import random

class Room:
    def __init__(self, initial_state):
        self.state = initial_state  # 1 indicates dirty, 0 indicates clean

    def clean(self, position):
        print(f"Cleaned position {position}.")
        self.state[position] = 0  # Mark the position as clean

    def display(self):
        print("Current Room State:", self.state)

class GoalBasedVacuumCleanerAgent:
    def __init__(self, room):
        self.room = room

    def clean_room(self):
        print("Starting cleaning process...")
        for position in range(len(self.room.state)):
            if self.room.state[position] == 1:  # Check if the room is dirty
                self.room.clean(position)  # Clean the dirty room
            else:
                print(f"Position {position} is already clean, skipping.")
        print("Cleaning process completed.")
        self.room.display()

# Initialize the room with a random state
initial_state = [random.randint(0, 1) for _ in range(10)]  # Create a room with 10 random states
room = Room(initial_state)
room.display()

# Create an agent and perform cleaning
agent = GoalBasedVacuumCleanerAgent(room)
agent.clean_room()
```

```python
import random

class Environment:
    def __init__(self, room_count):
        # Initialize room states randomly (1: dirty, 0: clean)
        self.room_count = room_count
        self.room_state = [random.randint(0, 1) for _ in range(room_count)]

    def get_dirty_positions(self):
        return [i for i, state in enumerate(self.room_state) if state == 1]

class ModelBasedVacuumAgent:
    def __init__(self, environment):
        self.environment = environment
        self.score = 0
        self.vacuum_position = 0  # Starting position
        self.history = []  # To keep track of the cleaned positions

        # Display the initial state
        print("Initial Room State (1 indicates dirty, 0 indicates clean):")
        print(self.environment.room_state)

        # Perform cleaning
        self.clean_rooms()

    def clean_rooms(self):
        while True:
            # Get the current dirty positions
            current_dirty_positions = self.environment.get_dirty_positions()

            if not current_dirty_positions:
                print("All rooms are clean.")
                break

            print(f"Current dirty positions: {current_dirty_positions}")

            # Clean the current position if it's dirty
            if self.environment.room_state[self.vacuum_position] == 1:
                self.clean_position(self.vacuum_position)

            # Move to the next position
            self.move_to_next_position()

    def clean_position(self, position):
        print(f"Cleaned position: {position}")
        self.environment.room_state[position] = 0  # Mark room as clean
        self.score += 1  # Increment score
        self.history.append(position)  # Log the cleaned position

    def move_to_next_position(self):
        # Move to the next position in a circular manner
        self.vacuum_position = (self.vacuum_position + 1) % self.environment.room_count
        print(f"Moved from {self.vacuum_position - 1} to {self.vacuum_position}")

# Create an environment and a model-based vacuum agent
room_count = 10  # Define the number of rooms
the_environment = Environment(room_count)
the_vacuum_agent = ModelBasedVacuumAgent(the_environment)
```

```python
import random

# Goal-based Agent
class GoalBasedThermostat:
    def __init__(self, desired_temperature):
        self.desired_temperature = desired_temperature

    def act(self, current_temperature):
        print(f"Current temperature: {current_temperature}°C.")
        if current_temperature < self.desired_temperature:
            print(f"Goal-based thermostat is heating the room to reach the desired temperature of {self.desired_temperature}°C.")
        elif current_temperature > self.desired_temperature:
            print(f"Goal-based thermostat is cooling the room to reach the desired temperature of {self.desired_temperature}°C.")
        else:
            print(f"Current temperature is already at the desired level of {self.desired_temperature}°C.")

# Model-based Agent
class ModelBasedThermostat:
    def __init__(self):
        self.temperature_model = {"outside_temperature": random.randint(-10, 35)}

    def perceive(self):
        # Simulate perceiving the outside temperature
        self.temperature_model["outside_temperature"] = random.randint(-10, 35)

    def act(self, current_temperature):
        self.perceive()
        outside_temp = self.temperature_model["outside_temperature"]
        print(f"Model-based thermostat perceives outside temperature: {outside_temp}°C.")
        print(f"Current temperature: {current_temperature}°C.")

        # Adjust temperature based on model
        if outside_temp < current_temperature:
            print("Model-based thermostat is cooling the room.")
        elif outside_temp > current_temperature:
            print("Model-based thermostat is heating the room.")
        else:
            print("Current temperature is stable; no action needed.")

# Utility-based Agent
class UtilityBasedThermostat:
    def __init__(self):
        self.actions = {
            "heat": 8,   # Heating has a utility value of 8
            "cool": 5,   # Cooling has a utility value of 5
            "maintain": 3  # Maintaining has a utility value of 3
        }

    def act(self, current_temperature, desired_temperature):
        print(f"Current temperature: {current_temperature}°C. Desired temperature: {desired_temperature}°C.")
        if current_temperature < desired_temperature:
            best_action = "heat"
        elif current_temperature > desired_temperature:
            best_action = "cool"
        else:
            best_action = "maintain"

        utility_value = self.actions[best_action]
        print(f"Utility-based thermostat chooses to '{best_action}' with utility value {utility_value}.")

# Main function to demonstrate the thermostats
def main():
    current_temperature = random.randint(10, 30)  # Random current temperature
    desired_temperature = 22  # Desired room temperature

    # Goal-based thermostat scenario
    print("=== Goal-Based Thermostat ===")
    goal_thermostat = GoalBasedThermostat(desired_temperature)
    goal_thermostat.act(current_temperature)

    print("\n=== Model-Based Thermostat ===")
    model_thermostat = ModelBasedThermostat()
    model_thermostat.act(current_temperature)

    print("\n=== Utility-Based Thermostat ===")
    utility_thermostat = UtilityBasedThermostat()
    utility_thermostat.act(current_temperature, desired_temperature)

if __name__ == "__main__":
    main()
```

```python
import random

class Area:
    def __init__(self, size, delivery_points):
        self.size = size
        self.area = self.generate_area(delivery_points)
        self.delivery_status = [False] * delivery_points

    def generate_area(self, delivery_points):
        return [1 if i < delivery_points else 0 for i in range(self.size)]

    def is_delivery_point(self, position):
        return self.area[position] == 1

    def deliver(self, position):
        if self.is_delivery_point(position) and not self.delivery_status[position]:
            self.delivery_status[position] = True
            print(f"Delivered item at position: {position}")

    def get_delivery_points(self):
        return [i for i in range(len(self.area)) if self.is_delivery_point(i) and not self.delivery_status[i]]

class ModelBasedDeliveryRobot:
    def __init__(self, area):
        self.area = area
        self.current_position = 0
        self.history = []

    def deliver_items(self):
        while True:
            delivery_points = self.area.get_delivery_points()
            print("Current undelivered positions:", delivery_points)
            if not delivery_points:
                print("All items delivered!")
                break
            next_position = delivery_points[0]
            self.move_to(next_position)
            self.area.deliver(next_position)

    def move_to(self, position):
        print(f"Moved from {self.current_position} to {position}")
        self.current_position = position
        self.history.append(position)

    def print_history(self):
        print("History of moves:", self.history)

area_size = 10
delivery_points_count = 5
area = Area(area_size, delivery_points_count)

print("Initial Area State (1 indicates delivery point, 0 indicates no delivery point):")
print(area.area)

delivery_robot = ModelBasedDeliveryRobot(area)
delivery_robot.deliver_items()

print("\nFinal Delivery Status:")
print(area.delivery_status)
delivery_robot.print_history()
```

```python
class Room:
    def __init__(self, initial_state):
        self.state = initial_state

    def extinguish(self, position):
        if self.state[position] == 1:
            self.state[position] = 0
            print(f"Extinguished fire at position {position}.")

    def display(self):
        print("Current Room State:", self.state)

class GoalBasedFireBrigadeAgent:
    def __init__(self, room, extinguish_goal):
        self.room = room
        self.extinguish_goal = extinguish_goal
        self.extinguished_count = 0

    def extinguish_fires(self):
        print("Starting fire extinguishing process...")
        for position in range(len(self.room.state)):
            if self.room.state[position] == 1 and self.extinguished_count < self.extinguish_goal:
                self.room.extinguish(position)
                self.extinguished_count += 1
            elif self.extinguished_count >= self.extinguish_goal:
                print(f"Reached extinguishing goal of {self.extinguish_goal} fires.")
                break
        print("Fire extinguishing process completed.")
        self.room.display()

initial_state = [1, 0, 1, 1, 0, 1, 1, 0, 0, 1]
room = Room(initial_state)
room.display()

extinguish_goal = 4
fire_brigade_agent = GoalBasedFireBrigadeAgent(room, extinguish_goal)
fire_brigade_agent.extinguish_fires()
```

```python
import random

class TemperatureRegulationAgent:
    def __init__(self):

        self.current_temperature = random.randint(0, 35)
        self.desired_ranges = {
            "comfortable": (20, 25),
            "warm": (25, 30),
            "cold": (15, 20),
            "too_hot": (30, 35),
            "too_cold": (0, 15)
        }
        self.utilities = {
            "comfortable": 10,
            "warm": 8,
            "cold": 6,
            "too_hot": 3,
            "too_cold": 2
        }

    def get_current_utilities(self):
        """Calculate the utility of the current temperature."""
        for range_name, (low, high) in self.desired_ranges.items():
            if low <= self.current_temperature <= high:
                return self.utilities[range_name]
        return 0

    def adjust_temperature(self):
        """Adjust the temperature based on utility values."""
        print(f"Current temperature: {self.current_temperature}°C")
        current_utility = self.get_current_utilities()
        print(f"Current utility value: {current_utility}")


        if current_utility < self.utilities["comfortable"]:
            if self.current_temperature < self.desired_ranges["comfortable"][0]:
                self.heating_on()
            elif self.current_temperature > self.desired_ranges["comfortable"][1]:
                self.cooling_on()
            else:
                print("Temperature is already comfortable.")
        else:
            print("No adjustments needed, temperature is within comfortable range.")

    def heating_on(self):
        print("Turning on the heating system to raise the temperature.")

    def cooling_on(self):
        print("Turning on the cooling system to lower the temperature.")


agent = TemperatureRegulationAgent()
agent.adjust_temperature()
```
